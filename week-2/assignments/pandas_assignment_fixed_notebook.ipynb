In this fifth task, we will focus on improving the readability and workability of the financial data within our dataset. Here, you'll need to identify the 'Value', 'Wage', and 'Release Clause' columns in the 'FIFA 21 Players' dataset, which contain '€', 'K', and 'M' characters.

**Inspecting the 'Value', 'Wage', and 'Release Clause' columns**

```python
fifa_df[['Value', 'Wage', 'Release Clause']].head()
```

We will remove the '€' character in these columns and convert the 'K' and 'M' suffixes in these fields as 'x1000' and 'x1,000,000' respectively. Lastly, the value will be converted to an 'int' data type.

```python
# Remove the '€' sign
fifa_df[['Value', 'Wage', 'Release Clause']] = fifa_df[['Value', 'Wage', 'Release Clause']].replace({'€': ''}, regex=True)

# Convert K and M sign to 1000 and 1000000
fifa_df[['Value', 'Wage', 'Release Clause']] = fifa_df[['Value', 'Wage', 'Release Clause']].replace({'K': 'e+03', 'M': 'e+06'}, regex=True).astype(float).astype(int)

fifa_df[['Value', 'Wage', 'Release Clause']]
```

**Identifying the players who have high 'Value' but relatively low 'Wage'**

We will define that players who have high 'Value' are players with 'Value' more than in the 75th percentile. While players who have relatively low 'Wage' are players with 'Wage' less than in the 25th percentile.

```python
wage_q1 = fifa_df['Wage'].quantile(0.25)
value_q3 = fifa_df['Value'].quantile(0.75)
fifa_high_val_low_wage = fifa_df[(fifa_df['Wage'] < wage_q1) & (fifa_df['Value'] > value_q3)]

# Player example
fifa_high_val_low_wage[['Name', 'Club', 'Wage', 'Value']].head(10)
```

**Creating a scatter plot from the filtered data**

```python
# @title #### 10. Scatter Plot Creation - High Value, Underpaid Players

# Put your code here:
plt.style.use('seaborn')

fifa_high_val_low_wage.plot(kind='scatter', x='Wage', y='Value', figsize=(15, 15), color='#343f85', s=75, alpha=0.75)
plt.xlabel('Wage (€)', labelpad=15, fontsize=20)
plt.ylabel('Value in M (€)', labelpad=15, fontsize=20)
plt.title('High Value Low Wage Players', fontsize=20, pad=15)

plt.savefig('high-value-underpaid-players.jpg', format='jpeg') # save image
plt.show()
# ---- End of your code ----

submit_image(student_id, '10. Scatter Plot Creation - High Value, Underpaid Players', './high-value-underpaid-players.jpg')
```
